```python
import io

df_player = pd.read_csv(io.BytesIO(uploaded['shai-gilgeous-alexander (1) (1).csv']))
#df_team = pd.read_csv(io.BytesIO(uploaded['league_four_factors_2024-02-02.csv']))
print(df_player)
```

```python
game_log_df = df_player[['GAME', 'OPP', 'MIN', 'REB', 'AST', 'TO', 'STL', 'PF', 'PTS']]
```

```python
game_log_df
```

```python
Y = game_log_df['PTS'].values
Y = Y.astype('int')
# Extract features (X) from the last 10 rows (excluding the first)
x = game_log_df.iloc[1:, 2:8]  # All columns except the target column pts

# Extract the target variable (y) for the most recent game
y = game_log_df.iloc[1:,8].values # Pts column
```

```python
x
```

```python
y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators= 1000, random_state = 42)
```

```python
model.fit(X_train, Y_train)
prediction_test = model.predict(X_test)
prediction_test
```

```python
# Check accuracy
acc = 0
acceptable_range = 3
for i in range(0, Y_test.size):
  acc += 0 if abs(Y_test[i]-prediction_test[i]) > acceptable_range else 1
acc /= Y_test.size
print("Accuracy = ", acc)
```
