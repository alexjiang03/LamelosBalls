```python
# Import the data
import pandas as pd
import numpy as np
vg_data = pd.read_csv('vgsales.csv')
vg_data
```

```python
# Remove rows missing an element
vg_data=vg_data.dropna()

# Replace strings with integers (function)
def fix_string_issue (col):
    key = []
    key = vg_data[col].unique()
        
    for i in range(0, len(vg_data[col].unique())): # For however many unique values of the column there are
        # Replace each instance with an integer assigned by how common it is
        vg_data.loc[vg_data[col] == vg_data[col].unique()[i], col] = i
    
    return key
        
plat_key = fix_string_issue('Platform')
gen_key = fix_string_issue('Genre')
pub_key = fix_string_issue('Publisher')
```

```python
# Input set: Platform, Genre, Publisher
X = vg_data.drop(columns = ['Rank', 'Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'])

# Output set: Global Sales
y = vg_data['Global_Sales']

# Fix issue with label type
from sklearn import preprocessing
lab_enc = preprocessing.LabelEncoder()
y = lab_enc.fit_transform(y)
```

```python
# Train model
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

# Split datasets into training & testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # Allocate 20% of dataset for testing.

def model_train ():
    # Split datasets into training & testing data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # Allocate 20% of dataset for testing.
    
    model.fit(X_train, y_train)

# Note model.predict([[Platform, Year, Genre, Publisher]])
```

```python
# Check accuracy
from sklearn.metrics import accuracy_score

def get_average_score ():
    scores = []
    def getscore ():
        while (1):
            model_train()
            predictions = model.predict(X_test)
            yield accuracy_score(y_test, predictions)

    gen = getscore()
    for i in range(0,20):
        scores.append(next(gen))

    avg_score = sum(scores)/len(scores)
    return avg_score
```

```python
# Saving and loading model
import joblib

def model_save ():
    joblib.dump(model, 'game-sales.joblib') # Save model to game-sales.joblib

def model_load():
    model = joblib.load('game-sales.joblib') # Load model from game-sales.joblib
```

```python
# Predict
# Note model.predict([[Platform, Year, Genre, Publisher]])

def model_prediction (Platform, Year, Genre, Publisher):
    Platform = np.where(plat_key == Platform)
    Genre = np.where(gen_key == Genre)
    Publisher = np.where(pub_key == Publisher)
    
    return model.predict([[Platform[0], Year, Genre[0], Publisher[0]]])[0]

for i in range (0,10):
    model_train()
    print(model_prediction('Wii', 2006, "Action", "Nintendo"))
```
